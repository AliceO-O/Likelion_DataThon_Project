```python
# import

import sys
import os

project_root = os.path.abspath(os.path.join("..", ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

print(project_root)

from src.naver_api import fetch_naver_items
from src.collector_naver_blog_news import (
    parse_blog_date,
    parse_news_date,
    count_naver_by_date,
    count_naver_unique_by_date,
    is_valid_naver_match,
    get_naver_unique_key
)
from datetime import datetime, timedelta
```

```python
# 블로그 raw 확인

blog_data = fetch_naver_items("blog", "REXTREME", start=1, display=5)
print("blog items 개수:", len(blog_data["items"]))
blog_data.keys()
```

```python
# 블로그 item별 검사(한국어)

for i, item in enumerate(blog_data["items"], start=1):
    print(f"[{i}]")
    print("date:", item.get("postdate"), "->", parse_blog_date(item))
    print("link:", get_naver_unique_key("blog", item))
    print("valid:", is_valid_naver_match(item, "렉스트림"))
    print("title:", item.get("title"))
    print("description:", item.get("description"))
    print("-" * 60)
```

```python
# 블로그 item별 검사(영어)

for i, item in enumerate(blog_data["items"], start=1):
    print(f"[{i}]")
    print("date:", item.get("postdate"), "->", parse_blog_date(item))
    print("link:", get_naver_unique_key("blog", item))
    print("valid:", is_valid_naver_match(item, "REXTREME"))
    print("title:", item.get("title"))
    print("description:", item.get("description"))
    print("-" * 60)
```

```python
# 뉴스 raw 확인

news_data = fetch_naver_items("news", "REXTREME", start=1, display=5)
print("news items 개수:", len(news_data["items"]))
news_data.keys()
```

```python
# 뉴스 item별 검사

for i, item in enumerate(news_data["items"], start=1):
    print(f"[{i}]")
    print("date:", item.get("pubDate"), "->", parse_news_date(item))
    print("unique_key:", get_naver_unique_key("news", item))
    print("valid:", is_valid_naver_match(item, "REXTREME"))
    print("title:", item.get("title"))
    print("description:", item.get("description"))
    print("-" * 60)
```

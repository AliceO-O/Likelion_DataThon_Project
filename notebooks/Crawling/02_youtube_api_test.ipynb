```python
from datetime import datetime, timedelta
from src.youtube_api import fetch_youtube_items
from src.collector_youtube import (
    make_utc_range_for_kst_day,
    count_youtube_by_date,
    is_valid_youtube_match,
    count_youtube_unique_by_date
)
```

```python
target_date = (datetime.now().date() - timedelta(days=1))
published_after, published_before = make_utc_range_for_kst_day(target_date)

print("target_date:", target_date)
print("published_after:", published_after)
print("published_before:", published_before)
```

```python
yt_data = fetch_youtube_items(
    query="REXTREME",
    published_after=published_after,
    published_before=published_before,
    max_results=5
)

print("검색 결과 개수:", len(yt_data["items"]))
yt_data.keys()
```

```python
for i, item in enumerate(yt_data["items"], start=1):
    snippet = item.get("snippet", {})
    title = snippet.get("title", "")
    description = snippet.get("description", "")
    valid = is_valid_youtube_match(item, "REXTREME")

    print(f"[{i}] valid={valid}")
    print("title:", title)
    print("description:", description[:100])
    print("-" * 60)
```

```python
import pandas as pd
```

```python
# 1) 파일 불러오기
df_main = pd.read_csv("../../data/Crawling_data/daily_counts.csv")
df_insta = pd.read_csv("../../data/Crawling_data/instagram_daily_counts(25.12.15~26.03.06).csv")
df_cafe = pd.read_csv("../../data/Crawling_data/naver_cafe_daily_counts.csv")
```

```python
df_insta.head(9)
```

```python
df_cafe.head(9)
```

```python
# 2) keyword 이름 통일
df_main["keyword"] = df_main["keyword"].replace({
    "combined_unique_url": "combined_unique"
})
df_cafe["keyword"] = df_cafe["keyword"].replace({
    "combined_unique_url": "combined_unique"
})
```

```python
df_main.head(9)
```

```python
df_cafe.head(9)
```

```python
# 3) 컬럼 순서 통일
cols = ["target_date", "platform", "keyword", "count", "collected_at", "status", "note"]

df_main = df_main[cols].copy()
df_insta = df_insta[cols].copy()
df_cafe = df_cafe[cols].copy()
```

```python
# 4) 세로 결합
df_tableau = pd.concat([df_main, df_insta, df_cafe], ignore_index=True)
```

```python
df_tableau.head(9)
```

```python
# 5) status 성공 데이터만 남기고 싶으면
df_tableau = df_tableau[df_tableau["status"] == "success"].copy()
```

```python
# 6) 날짜형 변환
df_tableau["target_date"] = pd.to_datetime(df_tableau["target_date"], errors="coerce")
```

```python
# 7) 정렬
platform_order = ["naver_blog", "naver_cafe", "naver_news", "youtube", "instagram"]
keyword_order = ["렉스트림", "REXTREME", "combined_unique"]

df_tableau["platform"] = pd.Categorical(
    df_tableau["platform"],
    categories=platform_order,
    ordered=True
)

df_tableau["keyword"] = pd.Categorical(
    df_tableau["keyword"],
    categories=keyword_order,
    ordered=True
)

df_tableau = df_tableau.sort_values(["target_date", "platform", "keyword"]).reset_index(drop=True)
```

```python
# 8) 저장
df_tableau.to_csv(
    "../../data/Crawling_data/tableau_all_platform_union.csv",
    index=False,
    encoding="utf-8-sig"
)

df_tableau.head(20)
```
